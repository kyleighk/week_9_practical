Q6. Use the `lm` function to model the richness as a function of the remaining variables but not including the variable `week`, which needs a special treatment we haven't covered yet!     


````r
richness_lm = lm(survey_richness ~ exposure + salinity + temperature + NAP + penetrability + grainsize + humus + chalk + sorting1 + Beach, survey_data_richness)
summary(richness_lm)
````

````r
survey_data_richness[1:45, c(10, 34)]
````

````r
richness_lm$coefficients
````

````r
ggplot(survey_data_richness) +
  geom_point(aes(x=NAP, y=survey_richness)) +
  geom_abline(intercept = richness_lm$coefficients[1], slope = richness_lm$coefficients[2], color = 2, size = 2, alpha = 0.5)
````

Q7. What do the various output of the `lm` mean? Interpret the results of your model. 

````r
# The residuals are the estimates of pop error. The difference between the y-values in the population and the predicted y-values in the model. 
# The coefficients are the weight that each parameter has on the model. Std error by how much error can occur and p-value aids in telling significance
# Residual standard error, deviation of residuals from model. 
# R^2  difference between residuals from the model, adjusted for sample size and variables. 
# F-stat if certain variables are significant, tells if a better fit is available

# In this case, the residuals have a median that is close to zero which would suggest a good fit but the 1 quartile and 3 quartile are not equal which might suggest uneveness in the model. The numbers should be randomly distributed. 
# Of the coefficients, only NAP had a significant p-value less than 0.05 and a standard error that was relatively low. 
# The residual standard error will be useful to plot and see how the residuals vary to get an idea of how the model fits the data and if it can be trusted. 
# The R-squared and adjusted R squared value are close-ish to 1. I think this means it is not a strong result but there is something significant. 
# The F-statistic reveals that there is a significant difference between column 10 and 34. 
````
